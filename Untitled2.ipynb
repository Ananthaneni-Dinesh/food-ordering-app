```python
class Food:
    id = 0
    
    def __init__(self, name, quantity, price, discount, stock):
        Food.id += 1
        self.food_id = Food.id
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

# Define a list of food items
food_list = []

# Define a function to add a new food item
def add_food_item(name, quantity, price, discount, stock):
    food = Food(name, quantity, price, discount, stock)
    food_list.append(food)
    return "Food item added successfully"


# Define a function to remove a food item
def remove_food_item(food_id):
    for food in food_list:
        if food.food_id == food_id:
            food_list.remove(food)
            return "Food item removed successfully"
    return "Food item not found"



# Define a function to edit a food item
def edit_food_item(food_id, name, quantity, price, discount, stock):
    for food in food_list:
        if food.food_id == food_id:
            food.name = name
            food.quantity = quantity
            food.price = price
            food.discount = discount
            food.stock = stock
            return "Food item edited successfully"
    return "Food item not found"

# Define a function to view all food items
def view_food_items():
    for food in food_list:
        print(f"Food ID: {food.food_id}")
        print(f"Name: {food.name}")
        print(f"Quantity: {food.quantity}")
        print(f"Price: {food.price}")
        print(f"Discount: {food.discount}")
        print(f"Stock: {food.stock}")

# Add new food items
add_food_item("Tandoori Chicken", "4 pieces", 240, 20, 50)
add_food_item("Vegan Burger", "1 Piece", 320, 10, 100)
add_food_item("Truffle Cake", "500gm", 900, 0, 75)
view_food_items()

# Edit a food item
print(edit_food_item(1, "Spicy Tandoori Chicken", "4 pieces", 280, 20, 50))
view_food_items()

# Remove a food item
print(remove_food_item(1))
view_food_items()



```
